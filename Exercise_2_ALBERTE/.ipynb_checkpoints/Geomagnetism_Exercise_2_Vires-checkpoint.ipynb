Loading the Swarm satellite B data through the ESA vires client:

```python
from viresclient import SwarmRequest # This is the package that allows us to query and download the Geomagnetic data.
import datetime as dt # This package allows us to describe the time limits for our queries.

request = SwarmRequest(); # Establishes communications with the Swarm Data Server.

request.set_collection("SW_OPER_MAGB_LR_1B"); # Telling the Swarm Data Server that we are interested in the Swarm Satellite B measurements.

request.set_products(                # Telling the Swarm Data Server the specific data we are interested in extracting;
    measurements = ["B_NEC"],        # We are interested in the X, Y, Z magnetic components measured by Swarm Satellite B.
    sampling_step = "PT01M",         # We want our queried data to consist of one sample per minute.
    auxiliaries = ["Kp10",           # We want to extract the KP10 index at the time of our Swarm measurements in order to filter out the non-quiet time data.
                   "Dst",            # We want to extract the DST index at the time of our Swarm measurements in order to filter out the non-quiet time data.
                   "SunZenithAngle"] # We want to extract the Sun Zenith Angle at time of our measurements in order to filter out sunlit data.
);

data = request.get_between(          # Telling the Swarm Data Server that our data should span from the 14'th of September 2018 to the 28'th of September 2018.
    start_time = dt.datetime(2018,9,14),
    end_time = dt.datetime(2018,9,29) # Note: This means the data will stop just before the 29'th (i.e. 23:59:00 on the 28'th).
);

df = data.as_dataframe(expand=True); # Turning our extracted Swarm data into a Pandas Dataframe (Easier to work with in Python). 
print(df.head()); # Printing the columns along with the first 5 rows of our extracted Swarm data, so we can see it.
```

## <center> Filtering the sunlit and quiet-time data </center>

Filtering the data by keeping all the data points with a time stamp corresponding to a solar zenith angle of 80 degrees or above.

```python
 # We specify that we are interested in keeping all data points where the corresponding Solar Zenith Angle is equal to or above 80 degrees.
Sunlit_Limit = df["SunZenithAngle"]>=80;

# We use the new boolean variable Sunlit_Limit to select all rows that lived up to our specification. 
#All the rows from the DST filtered data that lived up to the specification will then be saved as a new variable Data_Sunlit_Dst_Filtered.
Data_Sunlit_Filtered = df[Sunlit_Limit];

print('The length of the original data: ',len(df))
print('The length of the sunlit filtered data: ',len(Data_Sunlit_Filtered))
```

Filtering the data by keeping all the data points with a time stamp corresponding to a Kp10 index of 30 or below.

```python
# We specify that we are interested in keeping all data points where the KP10 index equal to or below 30 kp10 index.
Quiet_Time_Kp = Data_Sunlit_Filtered["Kp10"]<=30;

# We use the boolean variable Quiet_Time_Kp to select all rows that lived up to our KP10 specification.
# All the rows from the Sunlit and Dst_Filtered data that lived up to the Kp10 specification will then be saved as a final new variable Data_Sunlit_Dst_Kp_Filtered.
Data_Sunlit_Kp_Filtered = Data_Sunlit_Filtered[Quiet_Time_Kp];

print('The length of the sunlit & kp10 filtered data: ',len(Data_Sunlit_Kp_Filtered))
```

Filtering the data by removing all the data points with a time stamp corresponding to an hourly rate of change in the Dst index of 3 nT or above.

```python
# We convert all the measurements in our Dataframe to hourly means and save them as a new Dataframe (Hourly_DST) in order to obtain the Dst in hourly means instead of in minutes.
Hourly_DST = df.resample("1h").mean();

# We use the .diff function on the DST measurements in the new hourly mean Dataframe to find the difference between the DST in the current row and in the next row. 
# Then we look for any time the difference betweens the two rows (i.e. the two hours) is either bigger than 3 nT or smaller than 3 nT and save these in a new variable (Quiet_Time_DST_Limit).
# The Quiet_Time_DST_Limit variable gives us a list of all the hour intervals that need to be filtered out of our original dataset.
# The Quiet_Time_DST_Limit variable is a boolean variable where the noisy rows will be labeled as TRUE and the quiet rows will be labeled as FALSE.
Quiet_Time_DST_Limit = (Hourly_DST["Dst"].diff()>3) | (Hourly_DST["Dst"].diff()<-3);

# We use the new variable we just created (Quiet_Time_DST_Limit) to extract all the hours that were noisy. 
# In order to do this we use all the TRUE rows in Quiet_Time_DST_Limit to extract the corresponding measurements in our Hourly_DST dataframe via indexing.
Quiet_Time_DST_Filtered = Hourly_DST[Quiet_Time_DST_Limit];

# The hourly time stamps in the new variable we have made (Quiet_Time_DST_Filtered) are stored as index. 
# In order to use these time stamps to filter the noisy hours from our original dataset we store the noisy time index as a new variable (Noisy_DST_Hours) so it is no longer an index.
Noisy_DST_Hours = Quiet_Time_DST_Filtered.index;

# We now have all the noisy DST hours saved in the Noisy_DST_Hours variable.
# However, we need to grab all the measurements within these hours (i.e. if time stamp hour 15 is marked as noisy we wish to filter everything in-between hour 14 and 15. Not just the one time stamp at 15:00).

# Therefore, we create a new variable called Noisy_DST_Times that will grab all data points in-between the two noisy hourly DST measurements.
# In order to do this we match all the time stamps in the original data in-between the two noisy hours.
# To begin, we simply do this for the first noisy hour.
Noisy_Dst_Times = Data_Sunlit_Kp_Filtered.loc[Noisy_DST_Hours[0]-dt.timedelta(hours=1):Noisy_DST_Hours[0]]

# In order to do this for all the noisy DST hours we loop through them.
i=1;
while i <= len(Noisy_DST_Hours)-1:
    Noisy_Dst_Times = Noisy_Dst_Times.append(Data_Sunlit_Kp_Filtered.loc[Noisy_DST_Hours[i]-dt.timedelta(hours=1):Noisy_DST_Hours[i]])
    #print(df.loc[Times[i]-dt.timedelta(hours=1):Times[i]])
    i+=1

# At last we remove all the data in-between the noisy DST hours.
Data_Sunlit_Kp_Dst_Filtered = Data_Sunlit_Kp_Filtered.drop(Noisy_Dst_Times.index)
print('The length of the sunlit, Kp10 & Dstfiltered data: ' ,len(Data_Sunlit_Kp_Dst_Filtered))
print('The total amount of data that was filtered out of the original data set was: ', 100-len(Data_Sunlit_Kp_Dst_Filtered)/len(df)*100,'%')
```

## <center>Plotting the total magnetic intensity $F$ </center>

### <center> $F = \sqrt{X^2+Y^2+Z^2}$</center> 

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import cartopy.crs as ccrs

F = np.sqrt(np.square(Data_Sunlit_Kp_Dst_Filtered["B_NEC_N"].values)+np.square(Data_Sunlit_Kp_Dst_Filtered["B_NEC_E"].values)+np.square(Data_Sunlit_Kp_Dst_Filtered["B_NEC_C"].values))

############################################ Global Map ##############################################

f, ax = plt.subplots(1, subplot_kw=dict(projection=ccrs.EqualEarth()))

f.set_figheight(15);
f.set_figwidth(15);

h = ax.scatter(x=Data_Sunlit_Kp_Dst_Filtered["Longitude"].values, y=Data_Sunlit_Kp_Dst_Filtered["Latitude"].values, s = 40, c = F, edgecolors='none', cmap='plasma_r', transform = ccrs.PlateCarree());
ax.coastlines(linewidth=1.1);

divider = make_axes_locatable(ax);
ax_cb = divider.new_horizontal(size="5%", pad=0.1, axes_class=plt.Axes);

f.add_axes(ax_cb);

#colorbar and properties
cbar = plt.colorbar(h, cax=ax_cb);
cbar.set_label('Field intensity, [nT]', rotation=270, labelpad=25, fontsize='large');

ax.set_title('Global Map of observed F (nT) from Swarm data', weight='bold', fontsize='x-large');

############################################ Arctic Map ##############################################

f, ax = plt.subplots(1, subplot_kw=dict(projection=ccrs.Orthographic(0,90)))

f.set_figheight(7.5);
f.set_figwidth(7.5);

h = ax.scatter(x=Data_Sunlit_Kp_Dst_Filtered["Longitude"].values, y=Data_Sunlit_Kp_Dst_Filtered["Latitude"].values, s = 40, c = F, edgecolors='none', cmap='plasma_r', transform = ccrs.PlateCarree());
ax.coastlines(linewidth=1.1);

ax.set_title('Arctic Map of observed F (nT) from Swarm data', weight='bold', fontsize='x-large');

########################################### Antarctic Map ##############################################

f, ax = plt.subplots(1, subplot_kw=dict(projection=ccrs.Orthographic(180,-90)))

f.set_figheight(7.5);
f.set_figwidth(7.5);

h = ax.scatter(x=Data_Sunlit_Kp_Dst_Filtered["Longitude"].values, y=Data_Sunlit_Kp_Dst_Filtered["Latitude"].values, s = 40, c = F, edgecolors='none', cmap='plasma_r', transform = ccrs.PlateCarree());
ax.coastlines(linewidth=1.1);

ax.set_title('Antarctic Map of observed F (nT) from Swarm data', weight='bold', fontsize='x-large');
```

## <center> Converting the Geocentric NEC magnetic field components ($\vec{B}_{X}, \vec{B}_{Y}, \vec{B}_{Z}$) to spherical magnetic field components ($\vec{B}_{r}, \vec{B}_{\vartheta}, \vec{B}_{\lambda}$) </center>

### <center> $\vec{B}_{r} = -\vec{B}_{Z}$ , $\vec{B}_{\vartheta} = -\vec{B}_{X}$, $\vec{B}_{\lambda} = \vec{B}_{Y}$ </center>

```python
from scipy.io import savemat

# Making sure the time stamps on our data are saved as a column. Previously it only existed as an index and this won't be saved into the Matlab file.
# Saving them as strings because Matlab is unable to open them when saved in the default datetime64 format.
Data_Sunlit_Kp_Dst_Filtered["Time"] = np.datetime_as_string(Data_Sunlit_Kp_Dst_Filtered.index, unit='s');

Data_Sunlit_Kp_Dst_Filtered["B_r"] = -Data_Sunlit_Kp_Dst_Filtered["B_NEC_C"].values;
Data_Sunlit_Kp_Dst_Filtered["B_vartheta"] = -Data_Sunlit_Kp_Dst_Filtered["B_NEC_N"].values;
Data_Sunlit_Kp_Dst_Filtered["B_lambda"] = Data_Sunlit_Kp_Dst_Filtered["B_NEC_E"].values;

savemat("/home/sean/Documents/Geomagnetism_Ex2_Data_Output3.mat",{'struct':Data_Sunlit_Kp_Dst_Filtered.to_dict("list")})
```
