In the following notebook it will be demonstrated how to carry out data selection on typical Swarm data products, and plot some relevant geomagnetic features.  Uses pandas for data manipulation

Requires: numpy, matplotlib, pandas, cdflib


```python
import cdflib
import pandas as pd
import numpy as np
from datetime import datetime
import os
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from lib.solar_emphemeris import *
```

```python
# Define Paths and Files to read
DST_PATH = "Disturbance_Indices/Dst_MJD_1998.dat"
KP_PATH = "Disturbance_Indices/Kp_MJD_1998_QL.dat"
DATA_PATH = "Data/"
```

# (i) Load data

```python
# Load in Dst and Kp indices:
time_Dst, Dst = np.loadtxt(DST_PATH, usecols=(0,1), comments='#',unpack=True)
time_Kp, Kp = np.loadtxt(KP_PATH, usecols=(0,1), comments='#',unpack=True)

# Put into pandas dataframe for easy access
Dst_indices = pd.DataFrame({'time_Dst': time_Dst, 'Dst': Dst})
Kp_indices = pd.DataFrame({'time_Kp': time_Kp, 'Kp': Kp})
```

```python
# Load Swarm data
i = 0
mjd2000_time = []
radii = []
theta = []
phi = []
b_nec = []
flags_b = []
flags_q = []

for folder, subfolder, files in os.walk(DATA_PATH):
    for file in sorted(list(files)):
        # if there is any non-cdf files in your folder they will be skipped:
        try:
            
            cdf_file = cdflib.CDF(folder + file)
            time_stamps = cdf_file.varget("Timestamp")  # CDF epoch is in miliseconds since 01-Jan-0000
            print(folder + file)
            #tmp = (time_stamps - time_stamps[0]) / (1e3*60*60*24) + to_mjd2000(2018, 9, 14+i)
            mjd2000_time.extend((time_stamps - time_stamps[0]) / (1e3*60*60*24) + to_mjd2000(2018, 9, 14+i))  # 
            radii.extend(cdf_file.varget("Radius")/1e3)
            theta.extend(90 - cdf_file.varget("Latitude"))
            phi.extend(cdf_file.varget("Longitude"))
            b_nec.extend(cdf_file.varget("B_NEC"))
            flags_b.extend(cdf_file.varget("Flags_b"))
            flags_q.extend(cdf_file.varget("Flags_q"))
            i += 1
            cdf_file.close()
        except OSError:
            print('Error could not open file:', "\n", file)
            pass
```

```python
# Place the data in a pandas dataframe for easy data processing

b_nec = np.array(b_nec)
data = [mjd2000_time, radii, theta, phi, b_nec[:, 0], b_nec[:, 1], b_nec[:, 2], flags_b, flags_q]
column_names = ["time_stamp", "radius", "colat", "lon", "X", "Y", "Z", "flags_b", "flags_q"]

# place data in dataframe for easier nan-removal etc.
dataframe = pd.DataFrame()
for index, col in enumerate(column_names):
    dataframe[col] = data[index]

# save memory
del data, mjd2000_time, radii, theta, phi, b_nec, flags_q

print('shape of dataframe:', dataframe.shape)
```

```python
# drop nans, if any
dataframe = dataframe.dropna()

# check for error flags. Where flag_b or flag_q is 255

# drops rows where flag_b == 255
dataframe.drop(dataframe[dataframe.flags_b == 255].index, inplace=True)

# drops rows where flag_q == 255
dataframe.drop(dataframe[dataframe.flags_q == 255].index, inplace=True)
```

# (ii) Down-sample data

```python
# Take every 60th datum
dataframe = dataframe.iloc[::60, :]
dataframe.shape
```

# (iii) Exploratory plots of vector field components vs co-latitude and time

```python
# Plot the vector field components against co-latitude and time.
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(7,14))  # plotted in one figure

# subplot 1
ax[0].plot(dataframe.colat.values, dataframe.X.values, '.', label='X')
ax[0].plot(dataframe.colat.values, dataframe.Y.values, '.', label='Y')
ax[0].plot(dataframe.colat.values, dataframe.Z.values, '.', label='Z')

ax[0].grid()
ax[0].set_title("Dependence of Field Components on co-latitude, Swarm-B data", fontweight="bold")
ax[0].title.set_position([.5, 1.04])
ax[0].set_xlabel("co-latitude [deg]")
ax[0].set_ylabel("[nT]")
ax[0].legend()

# subplot 2
ax[1].plot(dataframe.time_stamp.values, dataframe.X.values, '.', alpha=0.5, label='X')
ax[1].plot(dataframe.time_stamp.values, dataframe.Y.values, '.', alpha=0.5, label='Y')
ax[1].plot(dataframe.time_stamp.values, dataframe.Z.values, '.', alpha=0.5, label='Z')

ax[1].grid()
ax[1].set_title("Dependence of Field Components on time, from Swarm data", fontweight="bold")
ax[1].title.set_position([.5, 1.04])
ax[1].set_xlabel("Time, MJD2000 [days]")
ax[1].set_ylabel("[nT]")
ax[1].legend()

plt.show()
```

# (iv) Remove Sunlight data

```python
# removing sunlit data
rad = np.pi/180
zenith = 90-10  # zenith angle 80 deg

# to compute how much data has reduced after selection:
n_data_all = dataframe.shape[0]

# to make cos_zeta expression more readable:
colat = dataframe.colat.values
lon = dataframe.lon.values
time = dataframe.time_stamp.values

# threshold for dark time observation
cos_zeta_0 = np.cos((zenith) * rad)
_, declination = sun_mjd2000(dataframe.time_stamp.values)
cos_zeta = np.cos(colat * rad) * np.sin(declination) \
         + np.sin(colat * rad) * np.cos(declination) * np.cos(np.mod(time + .5, 1) * 2*np.pi + lon * rad)

# rows with cos_zeta higher than cos_zeta_0 will be dropped.
sunlit_mask_inv = cos_zeta > cos_zeta_0
sunlit_mask = sunlit_mask_inv!=True

# select rows where sunlit_data_mask==True
dataframe = dataframe[sunlit_mask]

# save the number of data points after dark time selection
n_data_dark = dataframe.shape[0]

del colat, lon, time  # save memory
dataframe.shape
```

# (v) Implement quiet time selection based on rate of change of Dst and Kp

```python
# Quiet-time selection, using dDst and Kp thresholds
# normalize d_dst by difference in time in case two time rows have longer timestep than sample rate.
Dst=Dst_indices.Dst.values
d_dst = np.diff(Dst)
d_dst = np.hstack((d_dst[0], d_dst))  # so same size as Dst

# Interpolate in order to obtain d_Dst at times of interest:
d_dst_interp = np.interp(dataframe.time_stamp.values,  Dst_indices.time_Dst.values , d_dst)

# Interpolate in order to obtain Kp at times of interest:
Kp_interp = np.interp(dataframe.time_stamp.values, Kp_indices.time_Kp.values , Kp_indices.Kp)

# create boolean masks for data selection, exclude data when either condition is true met:
dst_Kp_inv_mask = (np.abs(d_dst_interp) > 3) | (Kp_interp > 30)
dst_Kp_mask = dst_Kp_inv_mask!=True   #invert mask

# select rows where dst_Kp_mask==True 
dataframe = dataframe[dst_Kp_mask]

# save the number of data points after dark and quiet time selection
n_data_dark_quiet = dataframe.shape[0]

print("Data remaining:", n_data_dark_quiet/n_data_all * 100, "%")
print("Final size:", dataframe.shape)

```

# (vi) Convert to field intensity F and plot vs co-latitude and longitude

```python
# Calculate field intensity and plot 

# field components
b_r = -dataframe.Z.values 
b_theta = -dataframe.X.values
b_phi = dataframe.Y.values

# intensity:
F = np.sqrt(dataframe.X.values**2 + dataframe.Y.values**2 + dataframe.Z.values**2)

from mpl_toolkits.axes_grid1 import make_axes_locatable

#create figure
f, ax = plt.subplots(1, 1, subplot_kw=dict(projection=ccrs.PlateCarree()))

# figure size
f.set_figheight(15) 
f.set_figwidth(15)

# cartopy map object
h = ax.scatter(x=dataframe.lon.values, y=90-dataframe.colat.values, s=40, c=F, edgecolors='none', transform=ccrs.PlateCarree(), cmap='plasma_r')
ax.coastlines(linewidth=1.1)

# we need to set axes_class=plt.Axes, else it attempts to create a GeoAxes as colorbar
divider = make_axes_locatable(ax)
ax_cb = divider.new_horizontal(size="5%", pad=0.1, axes_class=plt.Axes)

f.add_axes(ax_cb)

# colorbar and properties
cbar = plt.colorbar(h, cax=ax_cb)
cbar.set_label('Field intensity, [nT]', rotation=270, labelpad=25, fontsize='large')

ax.set_title('Map of observed F (nT) from Swarm data', weight='bold', fontsize='x-large')
```
